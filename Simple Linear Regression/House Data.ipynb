```python
# Predicting the price of a house from certain charactaristics
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
#Read in the dataframe 
df = pd.read_csv('kc_house_data.csv')
```

```python
#Visualize the data
print(df.head())


fig = plt.figure()
fig.set_figheight(18)
fig.set_figwidth(18)

#Visualize Each of the continous-valued columns
for idx,i  in enumerate([4,5,6,12,13]):
    plt.subplot(2,4,idx+1)
    plt.scatter(df.iloc[:,[i]],df['price'])
    plt.xlabel(df.columns.values[i])
    plt.ylabel('price')
```

```python
#Regression analysis of Single Column (sqft_living) on Price.

#Seperate into Testing and Training data
test = df['sqft_living'].values

#Edit Test
test = test.reshape((-1,1))
test = (test - test.mean())/test.std()
test = np.append(np.ones(len(test)).reshape((-1,1)),test,axis=1)

#Expected Output
y = df['price'].values
y = y.reshape((-1,1))

theta = np.transpose([0,0]).reshape(2,1)
```

```python
def cost_function(theta,test,y,m):
    cost = 1/(2*m) * np.sum(np.power((np.subtract(test.dot(theta),y)),2))
    return cost

alpha = .01 
num_iters = 2000
m = len(y)

theta = np.transpose([0,0]).reshape(2,1)
for i in range(1,num_iters):
    error = np.subtract(np.dot(test,theta),y)
    cost = 1/(2*m) * np.dot(error.T,error)
    theta = theta - (alpha /m * np.dot(test.T,error))
    print('Iteration: %s ,Theta Values: %s, Cost: %s'%(str(i),theta,str(cost)))

```

```python
print('Final Weights: %s'% str(theta))
```

```python
#Plot the regression line against the normalized data
fig = plt.figure()
fig.set_figheight(18)
fig.set_figwidth(18)

plt.scatter(test[:,1],y)
x = np.linspace(-5,10,1000)
plt.plot(x,theta[0,0]+ theta[1,0] * x,color = 'r')
plt.xlabel('sqft_living (normalized)')
plt.ylabel('price')
```

```python
# Perform the same task with sklearn
import scipy.stats as stats
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

lm.fit(test,y)

print('Theta Values from sklearn: %s,[%s]'%(lm.intercept_,lm.coef_[0,1]))
#Almost Exactly the same!!!!!!!!!
```

```python
fig = plt.figure()
fig.set_figheight(18)
fig.set_figwidth(18)

plt.scatter(y,lm.predict(test))

plt.ylabel('Predicted Prices')
plt.xlabel('Actual Prices')

plt.show()
```

```python
#Calculate the Mean Square Error: 
#Really high
np.mean(np.power(y-lm.predict(test),2))
```
